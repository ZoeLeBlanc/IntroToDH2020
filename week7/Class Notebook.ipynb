```python
film_scripts[(film_scripts.year == 2015)].sort_values(by=['gross_ia'], ascending=False)
```

```python
year_grossia = film_scripts.groupby('year')['gross_ia'].sum().reset_index()
```

```python
year_counts = film_scripts.groupby('year').size().reset_index(name='counts')
```

```python
film_scripts_merged = pd.merge(year_grossia, year_counts, on='year')
```

```python
import matplotlib as plt
```

```python
film_scripts_merged.plot.scatter(x='gross_ia', y='counts')
```
